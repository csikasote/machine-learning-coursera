## Introduction ##

Support Vector Machines (SVM) are a very powerful and versatile class of Machine Learning models, capable of performing linear or non-linear classification, regression, even outlier detection. They are particularly well suited for classification of complex but small or medium-sized datasets.

For classification, SVMs can be classified into two categories: 
- Linear SVM classification
- Nonlinear SVM classification

So lets discuss these one by one in the follow up sections:

### Linear SVM Classification ###

A dataset is said to be linearly separable if they can be separated by a straight line. With SVMs the objective is to maximise the `margin`. We can define a margin as the distance between the separating hyperplane (decision boundary) and the training samples that are closes to this hyperplane (also called `support vectors`). See the illustration in the figure below. This means we can think of SVM classification as fitting the widest possible street between two classes. This is referred to as `Large Margin Classification`. Therefore, adding more training instances "off the street" does not affect the decision boundary at all. 

![svm](images/svm.png)

With SVMs there are two margin classification: `soft margin classification` and `hard margin classification`.

- **Hard margin classification** is a type of margin classification where we all instances are supposed to be "off the street" and on the right side. This raises two concerns; only works if the data is linearly separable and it is sensitive to outliers.
- **Soft margin classification** is an alternative type of margin classification that seeks to address the issues with hard margin classification. The model seeks to find a good balance between keeping the street as large as possible and limiting the margin violations (i.e., instances that end up in gthe middle of the street or on the wrong side).

**Scikit-Learn Implementation of SVM for linear classification**

In Scikit-Learn's SVM classes, we control the hard vs soft margin classification balance using the C hyperparameter. 
- A smaller C value leads to a wider street but more margin violations. 
- Using a large C value leads to fewer margin violations but ends up with a smaller margin.

**Note:** If the SVM model is overfitting, we can attempt to reduce C for regularization.

Using Scikit-Learn for SVM linear classification, we can use the following classes:
- [LinearSVC](https://scikit-learn.org/stable/modules/generated/sklearn.svm.LinearSVC.html)
- [SVC with linear kernel](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html)
- [SDGClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDClassifier.html)

### Nonlinear SVM Classification ###

Despite that linear SVM classifiers are efficient and work very well in several cases, the reality is that many dataset are not linearly separable. it has been proposed that one way to handle nonlinear dataset  is to add more feature by way of polynomial features.

**Scikit-Learn Implementation of SVM for nonlinear classification**

Using scikit-learn for SVM nonlinear classification, we can use the following classes:

## Part 1 - SVM for Linear Classification ##

### 1.1 - Example Dataset 1 ###

We will begin with a 2D example dataset which can be separated by a linear boundary. In this dataset, the positions of the positive examples (indicated with +) and the negative examples (indicated with o) suggest a natural separation indicated
by the gap.

So let us load our first dataset `./data/ex6data1.mat` and visualize the data points.

```python
df_path = 'data/ex6data1.mat' 
dataset1    = sio.loadmat(df_path)
X1       = dataset1["X"];
y1       = dataset1["y"].ravel();
```

Now we implement a function `plot_dataset()` to plot the training dataset.

```python
def plot_dataset(X, y):
    #Find Indices of Positive and Negative Examples
    pos = (y == 1).nonzero(); 
    neg = (y == 0).nonzero();
    #plot the data
    plt.plot(X[pos, 0], X[pos, 1],'bs',linewidth=4, markersize=5, label='Positive');
    plt.plot(X[neg, 0], X[neg, 1],'g^',linewidth=4, markersize=5, label='Negative');
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)
    plt.grid(True)
```

The function `visualize_linear_boundary()` has been implemented in order to plot decision boundary as well as the suport vectors dividing the hyperplane.

```python
def visualize_linear_boundary(model,X,y, ax= None):
    ax = ax or plt.gca()
    w = model.coef_[0];
    b = model.intercept_[0];
    
    # At the decision boundary, w0*x0 + w1*x1 + b = 0
    # => x1 = -w0/w1 * x0 - b/w1
    xmin = X.min(); xmax = X.max();
    xp = np.linspace(xmin, xmax, 200)
    yp = -w[0]/w[1] * xp - b/w[1]
    
    margin = 1/w[1]
    gutter_up = yp + margin
    gutter_down = yp - margin
    
    svs = model.support_vectors_
    plt.scatter(svs[:, 0], svs[:, 1], s=180, linewidth=8, facecolors='#FFAAAA')
    plt.plot(xp, yp, "k-", linewidth=2)
    plt.plot(xp, gutter_up, "k--", linewidth=2)
    plt.plot(xp, gutter_down, "k--", linewidth=2)
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)
    plot_dataset(X, y)
```

For our dataset 1 in this exercise, we shall use Scikit-learn's `SVC` parameterised by `linear kernel` and hyperparameter `C` to implement SVM for linear classification. To implement this, we create an instance of SVM and fit our data since we wish to also plot the support vectors.

```python
C = 1.0  # SVM regularization parameter
svm_linear_kernel_model = SVC(kernel="linear", C=C) # SVM Classifier model
svm_linear_kernel_model = svm_linear_kernel_model.fit(X1, y1)
```

Next we plot the **dataset** as well as the `SVC with Linear Kernel` ..

```python
plt.figure(figsize=(15, 4))
plt.subplot(1, 2, 1)
plot_dataset(X1, y1)
plt.annotate("Outlier",
             xy=(0.1, 4.1),
             xytext=(1.0, 4.5),
             ha="center",
             arrowprops=dict(facecolor='black', width=2, shrink=0.1),
             fontsize=14,
            )
plt.title('Example Dataset 1')
plt.axis([0,5,1.5,5])

plt.subplot(1, 2, 2)
visualize_linear_boundary(svm_linear_kernel_model,X1,y1)
plt.annotate("Outlier",
             xy=(0.1, 4.1),
             xytext=(0.4, 3.0),
             ha="center",
             arrowprops=dict(facecolor='black', width=2, shrink=0.1),
             fontsize=14,
            )
plt.title('SVC with Linear Kernel (C = %d)'%(1))
plt.axis([0,5,1.5,5])
plt.show()
```

### 1.2 - Varying C to observe the behaviour of SVM towards outliers ###

It has been said that SVMs are sensitive to outliers. You will notice that there is an outlier positive example +(blue) on the far left at about (0.1, 4.1) in our dataset. As part of this exercise, we will attempt to see how this outlier affects the SVM decision boundary. So, we will try using different values of the C parameter with SVMs. Recall that the C parameter is a positive value that controls the penalty for misclassified training examples. A large C parameter tells the SVM to try to classify all the examples correctly.

```python
Cvals= np.array([1,100])

plt.figure(figsize=(len(Cvals) * 3 + 6, 4))
plt.subplots_adjust(left=.02, right=.98, bottom=.001, top=.96, wspace=.2, hspace=.1)

for idx, Cval in enumerate(Cvals):
    plt.subplot(1, len(Cvals), idx+1)
    model = SVC(kernel="linear", C=Cval).fit(X1, y1)
    visualize_linear_boundary(model,X1,y1)
    plt.annotate("Outlier",
             xy=(0.1, 4.1),
             xytext=(0.5, 3.0),
             ha="center",
             arrowprops=dict(facecolor='black', width=2, shrink=0.1),
             fontsize=14,
            )
    plt.title('SVM with Linear Kernel (C = %d )'%(Cval))
    plt.axis([0,5,1.5,5])
    idx += 2
plt.show()
```

We can see from the plots that:
- when C = 1, there is a wider street with more margin violations (i.e., instances can be seen inside the street) and the SVM puts the decision boundary in the gap between the two datasets and misclassifies the data point on the far left (outlier point).
- when C = 100, there is a smaller margin with few margin violations and the SVM now classifies every single example correctly, but has a decision boundary that does not appear to be a natural fit for the data

### 1.3 - Understanding decision functions & predictions ###

The Linear SVM classifier model predicts the class of a new instance by computing the decision function given by:

$$ h = {W^T.X + b} = w_1x_1 + w_2x_2 + w_3x_3 +....+ w_nx_n + b$$

If the result is positive, the predicted class $\hat y$ is the positive class(1), or else it is the negative class(0).

In order to get the intuition of decisions and predictions, let us plot the 3D of our trained model, `svm_linear_kernel_model`. 

```python
from mpl_toolkits.mplot3d import Axes3D

def plot_3D_decision_function(ax, model, x1_lim=[0, 5], x2_lim=[0, 5]):
    w, b, svs=model.coef_[0], model.intercept_[0], model.support_vectors_    
    x1_in_bounds = (X1[:, 0] > x1_lim[0]) & (X1[:, 0] < x1_lim[1])
    X_crop = X1[x1_in_bounds]
    y_crop = y1[x1_in_bounds]
    x1s = np.linspace(x1_lim[0], x1_lim[1], 20)
    x2s = np.linspace(x2_lim[0], x2_lim[1], 20)
    x1, x2 = np.meshgrid(x1s, x2s)
    xs = np.c_[x1.ravel(), x2.ravel()]
    df = (xs.dot(w) + b).reshape(x1.shape) # decision function
    m = 1 / np.linalg.norm(w)
        
    boundary_x2s = -x1s*(w[0]/w[1])-b/w[1]
    margin_x2s_1 = -x1s*(w[0]/w[1])-(b-1)/w[1] # gutter_up
    margin_x2s_2 = -x1s*(w[0]/w[1])-(b+1)/w[1] # gutter_down
    ax.plot_surface(x1s, x2, np.zeros_like(x1),color="b", alpha=0.3, cstride=100, rstride=100)
    ax.plot(x1s, boundary_x2s, 0, "k-", linewidth=2, label=r"$h=0$")
    ax.plot(x1s, margin_x2s_1, 0, "k--", linewidth=2, label=r"$h=\pm 1$")
    ax.plot(x1s, margin_x2s_2, 0, "k--", linewidth=2)
    ax.scatter(svs[:, 0], svs[:, 1], s=100, linewidth=3, facecolors='#FFAAAA')
    ax.plot(X_crop[:, 0][y_crop==1], X_crop[:, 1][y_crop==1], 0, "bs")
    ax.plot_surface(x1s, x2, df,color="y", alpha=0.1, cstride=100, rstride=100)
    ax.plot_wireframe(x1, x2, df, alpha=0.3, color="k")
    ax.plot(X_crop[:, 0][y_crop==0], X_crop[:, 1][y_crop==0], 0, "g^")
    ax.axis(x1_lim + x2_lim)
    ax.text(2.5, 4.0, 7.0, "d-function", fontsize=12)
    ax.set_xlabel(r"$X_1$", fontsize=15)
    ax.set_ylabel(r"$X_2$", fontsize=15)
    ax.set_title("Decision function for the training dataset", loc='center')
    ax.set_zlabel(r"$h = \mathbf{w}^T \mathbf{x} + b$", fontsize=14)
    #ax.legend(loc="upper left", fontsize=16)
```

```python
fig = plt.figure(figsize=(15, 4))

ax1 = fig.add_subplot(121)
visualize_linear_boundary(svm_linear_kernel_model,X1,y1)
plt.title('SVC with Linear Kernel (C = %d)'%(C))
plt.axis([0,5,1.5,5])

ax2 = fig.add_subplot(122, projection='3d')
plot_3D_decision_function(ax2, svm_linear_kernel_model)
plt.show()
```

The 3D plot above depicts the decision function that corresponds to the model we trained and plotted in 2D. It is a two-dimensional plane since the dataset has two features. 

The decision boundary for this is the set of points where the decision function is equal to 0. The intersection of two planes, which is a straight line (represented by the thick solid line). The dashed lines represent the points where the decision is equal to 1 or -1. They are parallel and at equal distance to the decision boundary, forming a margin around it.

Training a linear SVM classifier means finding the value of `W` and `b` that make this design as wide as possible while avoiding margin violations (hard margin) or limiting them (soft margin).

## Part 2 - SVM for Nonlinear Classification ##

### 2.1 - Example Dataset 2###

In this part of the exercise, you will be using SVMs to do non-linear classification. In particular, we will be using SVMs with Gaussian kernels on datasets that are not linearly separable.

As usual, we load and visualize our second 2D dataset, `./data/ex6data2.mat`. 

```python
df_path  = 'data/ex6data2.mat' 
Dataset2 = sio.loadmat(df_path)
X2       = Dataset2["X"];
y2       = Dataset2["y"].ravel();
```

```python
def plot_predictions(clf,X,y):  
    x0s = np.linspace(X.min()-0.05, X.max()+0.05, 100)
    x1s = np.linspace(y.min()+0.38, y.max()+0.01, 100)

    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]   
    y_pred = clf.predict(X).reshape(x0.shape)
    y_decision = clf.decision_function(X).reshape(x0.shape)
    
    plt.figure(figsize=(18,5))
    plt.subplots_adjust(wspace=0.4, hspace=0.4)
    
    plt.subplot(121)
    plot_dataset(X2, y2)
    plt.title("Example Dataset 2", fontsize=18)
    
    plt.subplot(122)
    plt.title("SVC with Gaussian Kernel", fontsize=18)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.1)
    plot_dataset(X2, y2)
```

As discussed in the introduction, the idea behind `kernel methods` to deal with linearly inseparable data is to create nonlinear combination of the original features to project them onto a higher dimensional space via a mapping function where it becomes linearly separable.

We can think of the Gaussian kernel as a similarity function that measures the "distance" between a pair of examples,
${(x^{(i)}, x^{(j))}}$. The Gaussian kernel is also parameterized by a bandwidth parameter, $\sigma$, which determines how fast the similarity metric decreases (to 0) as the examples are further apart.

The `Gaussian kernel`  or `Radial Basis Function (RBF) kernel` is defined as: 

![Gaussian_Kernel](images/gaussian_kernel.png)

This is often simplified as:

$$K_{gaussian}\,{(x^{(i)}, x^{(j)})} = {exp\,\biggl(-\gamma \parallel x^{(i)} - x^{(j)} \parallel \biggl)}$$

where, $\gamma = \frac {1}{2\sigma^ {2}}$ and the minus sign inverts th distance measure into a similarity score.

Now that we have an idea behind the kernel trick, let us see if we can train a kernel SVM that is able to draw a nonlinear decision boundary using Scikit-Learn's SVC class parameterised by `'rbf'` kernel, `gamma` and `C`.

For our implementation we shall compute the  gamma $\gamma$ value based on the $\sigma$. So we set $\sigma = 0.1$ value and **C = 100** as per the formula for calculating $\gamma$.

```python
sigma = 0.1;             # sigma component of the gaussian kernel formula
gamma = 0.5 * sigma**-2; # Computing gamma for gaussian kernal
rbf_kernel_svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="rbf", gamma=gamma, C=1.0))
    ])
rbf_clf = rbf_kernel_svm_clf.fit(X2, y2)
plot_predictions(rbf_clf,X2,y2)
```

The figures above shows the decision boundary found by the SVM with a Gaussian kernel. The decision boundary is able to separate most of the positive and negative examples correctly and follows the contours of the dataset well.

### 2.2 - Example Dataset 3 ###

In this part of the exercise, we will gain more practical skills on how to use a SVM with a Gaussian kernel. So let us load and visualize dataset 3, `./data/ex6data3.mat`.

```python
df_path  = 'data/ex6data3.mat' 
Dataset3       = sio.loadmat(df_path)
X3       = Dataset3["X"];
y3       = Dataset3["y"].ravel();
Xval     = Dataset3["Xval"];
yval     = Dataset3["yval"].ravel();
```

Our task is to use the cross validation set `Xval`, `yval` to determine the best C and $\sigma$ parameter to use. For both C and $\sigma$ , we shall try values in multiplicative steps (e.g., 0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30).

```python
def cross_validation(X3,y3,Xval,yval,val_range):
    pair  = (0, 0) 
    score = 0
    for c in val_range:
        for sigma in val_range:
            gamma = 0.5 * sigma**(-2);
            rbf_kernel_svm_clf = Pipeline([
                    ("scaler", StandardScaler()),
                    ("svm_clf", SVC(kernel="rbf", gamma=gamma, C=c))
                ])
            rbf_model = rbf_kernel_svm_clf.fit(X3, y3)
            cur_score = rbf_model.score(Xval,yval)
            if cur_score > score:
                score = cur_score
                pair = (c, sigma)
    return pair[0],pair[1],score
```

```python
def plot_prediction(clf,X,y):  
    x0s = np.linspace(X.min(), X.max()-0.25, 100)
    x1s = np.linspace(y.min()-0.7, y.max()-0.4, 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]   
    y_pred = clf.predict(X).reshape(x0.shape)
    y_decision = clf.decision_function(X).reshape(x0.shape)
    
    plt.figure(figsize=(15,4))
    plt.subplots_adjust(wspace=0.4, hspace=0.4)
    
    plt.subplot(121)
    plot_dataset(X3, y3)
    plt.title("Example Dataset 3", fontsize=18)
    
    plt.subplot(122)
    plt.title("SVC with Gaussian Kernel", fontsize=18)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.1)
    plot_dataset(X3, y3)
```

Now let us run a and evaluate our cross-validation function on a list of val_range[0.01, 0.03, 0.1, 0.3, 1., 3., 10., 30.]

```python
val_range = [0.01, 0.03, 0.1, 0.3, 1., 3., 10., 30.]
C, sigma,score = cross_validation(X3,y3,Xval,yval,val_range)
print('\nCross-validation completed ....')
print("\nBest (C, sigma) pair found by cross-validation is (C = %.2f, sigma = %.2f) with a score of %.4f.\n"%(C,sigma,score))
```

```python
gamma = 0.5 * sigma**-2;
rbf_kernel_svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="rbf", gamma=gamma, C=C))
    ])
rbf_model = rbf_kernel_svm_clf.fit(X3, y3)
plot_prediction(rbf_model,X3,y3)
```

For our best parameters, C= 1.0 and $\sigma$ = 0.3, the SVM returned a decision boundary shown in the figure on the right side!
